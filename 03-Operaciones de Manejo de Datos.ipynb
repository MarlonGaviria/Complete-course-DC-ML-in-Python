### <font color = green> Generacion de numeros aleatorios

```python
import numpy as np
```

```python
## funcion para albergar varios numeros aleatorios
def aleatorio(n,a,b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
aleatorio(25,1,50)
```
